```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../data/2020_2021_nba_season_player_stats.csv')
```

```python
df = df[~df['Unnamed: 0'].isna()]
df = df.drop(columns=['Unnamed: 0'])
df['game_location'] = df['game_location'].replace(['@', np.nan], ['Away', 'Home'])
df
```

```python
list(df.columns)
```

```python
df = df.rename(columns={'team_id': 'team_name', 'opp_id': 'opponent_name', 'gs': 'started_game', 'mp': 'minutes_played', 'fg': 'field_goals_made', 'fga': 'field_goals_attempts', 'fg_pct': 'field_goal_percentage', 'fg3':'field_goals_three_made', 'fg3a': 'field_goals_three_attempts', 'fg3_pct': 'field_goal_three_percentage', 'ft': 'free_throws_made', 'fta': 'free_throw_attempts', 'ft_pct': 'free_throw_percentage', 'orb': 'offensive_rebounds', 'drb':'defensive_rebounds', 'trb': 'total_rebounds', 'ast': 'assists', 'stl': 'steals', 'blk': 'blocks', 'tov': 'turn_overs', 'pf': 'personal_fouls', 'pts': 'points_scored', 'ranker': 'team_season_game', 'game_season': 'player_season_game'})
```

```python
df['started_game'] = df['started_game'].replace(['0', '1'], ['No', 'Yes'])
df
```

```python
#df[df.isna().any(axis=1)]
df.isna().any()
```

```python
df['field_goal_percentage'] = df['field_goal_percentage'].replace([np.nan], [-1])
df['field_goal_three_percentage'] = df['field_goal_three_percentage'].replace([np.nan], [-1])
df['free_throw_percentage'] = df['free_throw_percentage'].replace([np.nan], [-1])
df
```

```python
df[df['minutes_played'].isna()]
```

```python
df.drop(df[df['minutes_played'].isna()].index, inplace=True)
```

```python
df
```

```python
df.isna().any()
```

```python
df[df['plus_minus'].isna()]
```

```python
df.drop(df[df['plus_minus'].isna()].index, inplace=True)
```

```python
df.isna().any()
```

```python
df.to_csv('../data/clean_2020_2021_nba_season_player_stats.csv', index=False)
```
